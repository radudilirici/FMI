```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.cos(2 * x) - 2 * np.sin(3 * x)
```

```python
st, dr = -np.pi, np.pi
nr = 100
N = 10

x_gr = np.linspace(st, dr, nr)
y_gr = f(x_gr)

plt.plot(x_gr, y_gr, label='f(x)')

x = np.linspace(st, dr, N+1)
y = f(x)


def functie_spline_liniara(i):
    a = y[i]
    b = (y[i+1] - y[i]) / (x[i+1] - x[i])
    
    def spline(t):
        return a + b * (t - x[i])
    
    return spline


y_aprox = np.piecewise(
    x_gr,
    [
        (x[i] <= x_gr) & (x_gr < x[i + 1])
        for i in range(N - 1)
    ],
    [
        functie_spline_liniara(i)
        for i in range(N)
    ]
)

plt.plot(x_gr, y_aprox, color='orange', linestyle='--', label=f'aproximare')

plt.legend()
plt.show()
```
