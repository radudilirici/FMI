```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.cos(2 * x) - 2 * np.sin(3 * x)

def P(x, a):
    return sum([ai * x ** i for i, ai in enumerate(a)])
```

```python
def rezolva(f, st, dr, N=10):
    nr = 100
    x_gr = np.linspace(st, dr, nr)
    y_gr = f(x_gr)
    plt.plot(x_gr, y_gr, label='f(x)')
    
    X = np.linspace(st, dr, N + 1)
    y = f(X)
    plt.scatter(X, y, color='black')
    
    A = np.vander(X, N+1, increasing=True)
    b = y.reshape((-1, 1))
    a = np.linalg.solve(A, b)
    
    y_p = P(x_gr, a)
    plt.plot(x_gr, y_p, color='orange', linestyle='--', label=f'P(x) cu {N+1} puncte')
    
    plt.legend()
    plt.show()
```

```python
rezolva(f, -np.pi, np.pi, 6)
```
