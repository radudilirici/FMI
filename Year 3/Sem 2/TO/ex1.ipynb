```python
import numpy as np
from matplotlib import pyplot as plt
```

### Programe liniare:

```python
x = np.linspace(-2,10,100)
fig, ax = plt.subplots()

ax.plot(x, 4 + x)
ax.plot(x, 8 - x)
ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.fill([0,0,2,8], [0,4,6,0], 'green', alpha=0.5)
plt.plot(2,6, 'ro', markersize=10, alpha=0.8)

plt.xlim(-1, 9)
plt.ylim(-1, 7)

print("Solutie unica")
plt.show()
```

### Compatibil cu solutie unica:

de maximizat:

`x1 + 2 * x2`


conditii:

`- x1 + x2 <= 4`

`x1 + x2 <= 8`

`x1, x2 >= 0`

```python
x = np.linspace(-2,10,100)
fig, ax = plt.subplots()

ax.plot(x, 4 + x)
ax.plot(x, (8 - x) / 2)
ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.fill([0,8,0], [4,0,0], 'green', alpha=0.5)
x = np.linspace(0,8,100)
ax.plot(x, (8 - x) / 2, 'r', linewidth=5)

plt.xlim(-1, 9)
plt.ylim(-1, 9)

print("Infinitate de solutii")
plt.show()
```

### Compatibil cu o infinitate de solutii:

de maximizat:

`x1 + 2 * x2`


conditii:

`- x1 + x2 <= 4`

`x1 + 2 * x2 <= 8`

`x1, x2 >= 0`

```python
x = np.linspace(-2,10,100)
fig, ax = plt.subplots()

ax.plot(x, 4 + x)
ax.plot(x, (8 - x) / 2)
ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.fill([0,8,10,10], [4,0,0,14], 'green', alpha=0.5)

plt.xlim(-1, 9)
plt.ylim(-1, 9)

print("Optim infinit")
plt.show()
```

### Compatibil cu o optim infinit:

de maximizat:

`x1 + 2 * x2`


conditii:

`- x1 + x2 <= 4`

`x1 + 2 * x2 >= 8`

`x1, x2 >= 0`

```python
x = np.linspace(-2,10,100)
fig, ax = plt.subplots()

ax.plot(x, (2 - x) / 2)
ax.plot(x, (8 - x) / 2)
ax.grid(True, which='both')

ax.axhline(y=0, color='k')
ax.axvline(x=0, color='k')

plt.fill([0,8,10,10,0], [4,0,0,14,14], 'orange', alpha=0.5)
plt.fill([0,2,0], [1,0,0], 'blue', alpha=0.4)

plt.xlim(-1, 9)
plt.ylim(-1, 9)

print("Incompatibil")
plt.show()
```

### Incompatibil:

de maximizat:

`x1 + 2 * x2`


conditii:

`x1 + 2 * x2 <= 2`

`x1 + 2 * x2 >= 8`

`x1, x2 >= 0`
